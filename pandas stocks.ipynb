```python
df[['0053','0056','2303','2330','2317']].plot()
```

```python
stocks=['0050','0056','2303','2330','2454','2881','2542','00662']
```

```python
df[stocks].plot()
```

```python
df1=df[stocks].dropna()
```

```python
df1.head()
```

```python
df2=df1['2020-01-01':'2022-03-20']
```

```python
df2.plot(figsize=(10,5))
```

```python
((df2.pct_change()+1).cumprod()).plot(figsize=(10, 7))
```

```python
import matplotlib.pyplot as plt

((df2.pct_change()+1).cumprod()).plot(figsize=(10, 7))
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
